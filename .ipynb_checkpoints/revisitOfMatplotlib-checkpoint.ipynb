```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi) #0から2πまで50刻み
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.xlabel('x value')
plt.ylabel('y value')

plt.title('sin/cos')

plt.plot(x, y_sin, label='sin')
plt.plot(x, y_cos, label='cos', linestyle="dashed")
plt.legend() #凡例

plt.show()
```
