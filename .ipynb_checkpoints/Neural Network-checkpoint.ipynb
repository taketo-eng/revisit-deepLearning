```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2>神経細胞のモデル化</h2>
<p>コンピュータ上のモデル化された神経細胞ネットワークのことを人工ニューラルネットワーク(ANN, Artificial Neural Network)という</p>

<p>ニューロンには複数の入力があるが、出力は1つ</p>
<p>各入力には<b>重み</b>を掛け合わせる(重みは結合荷重ともいう)</p>
<p>-> 入力ごとに値が異なる</p>
<p><br>入力と重みを掛け合わせた値の総和に<b>バイアス</b>と呼ばれる定数を足す<br>(バイアス：ニューロンの感度で、大小によって興奮のしやすさが調整される)</p>

<p>この値を活性化関数で処理し、興奮状態を表す信号に変換する</p>

x:ニューロンへの入力値、w:重み
$$
xw
$$

n個の入力値と重さの積の総和

$$
\sum_{k=1}^nx_kw_k
$$

バイアスを加える

$$
u=\sum_{k=1}^n(x_kw_k)+b
$$

活性化関数をfとすると

$$
y=f(u)=f(\sum_{k=1}^n(x_kw_k)+b)
$$

<h2>ニューロンのネットワーク化</h2>
<p>ニューラルネットワークにおける層は、<b>入力層、中間層（隠れ層）、出力層の3つに分類可能</b></p>

<p>ニューロンの演算処理は<b>中間層と出力層のみで行われる</b><br>入力層は中間層に値を渡すだけ</p>

<p>入力から出力に向けて情報が流れることを順伝播、逆向きに遡っていくことを逆伝播という</p>

<h2>活性化関数</h2>
<h3>ステップ関数</h3>
<p>ニューラルネットワークの起源であるパーセプトロンで用いられていた</p>

$$
y=
\begin{cases}
0 \quad x \leqq 0 \\
1 \quad x > 0 \\
\end{cases}
$$

```python
def step_function(x):
    return np.where(x<=0, 0, 1)
```

```python
x = np.linspace(-5, 5)
y = step_function(x)
```

```python
plt.plot(x, y)
plt.show()
```

<h3>シグモイド関数</h3>
ステップ関数に比べて滑らかで、微分が扱いやすい
$$
y=\frac{1}{1+exp(-x)}
$$

```python
def sigmoid_function(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-5, 5)
y = sigmoid_function(x)
```

```python
plt.plot(x, y)
plt.show()
```

<h3>tanh:ハイパボリックタンジェント(hyperbolic tangent)</h3>
<p>0を中心とした対象になっておりバランスのいい活性化関数</p>
$$
y=\frac{exp(x)-exp(-x)}{exp(x)+exp(-x)}
$$

```python
def tanh_function(x):
    return np.tanh(x)
```

```python
x = np.linspace(-5, 5)
y = tanh_function(x)
```

```python
plt.plot(x, y)
plt.show()
```

<h3>ReLU</h3>
$$
y=
\begin{cases}
0 \quad x \leqq 0 \\
x \quad x > 0 \\
\end{cases}
$$

```python
def relu_function(x):
    return np.where(x <= 0, 0, x)
```

```python
x = np.linspace(-5, 5)
y = relu_function(x)
```

```python
plt.plot(x, y)
plt.show()
```

<h3>Leaky ReLU</h3>
$$
y=
\begin{cases}
0.01x \quad x \leqq 0\\
x \quad x > 0
\end{cases}
$$

```python
def leaky_relu_function(x):
    return np.where(x <= 0, 0.01*x, x)
```

```python
x = np.linspace(-5, 5)
y = leaky_relu_function(x)
```

```python
plt.plot(x, y)
plt.show()
```

<h3>恒等関数</h3>
$$
y=x
$$

```python
x = np.linspace(-5, 5)
y = x
```

```python
plt.plot(x,y)
plt.show()
```

<h3>ソフトマックス関数</h3>
<p>分類問題を扱うのに適した活性化関数</p>
$$
y=\frac{exp(x)}{\sum_{k=1}^nexp(x_k)}
$$

```python
def softmax_function(x):
    return np.exp(x)/np.sum(np.exp(x))
```

```python
x = np.linspace(-5, 5)
y = softmax_function(x)
```

```python
plt.plot(x,y)
plt.show()
```

<h2>ニューラルネットワークの実装</h2>

<h3>単一のニューロンを実装</h3>
<p>x, y座標の2つを入力として、出力を計算</p>

```python
X = np.arange(-1.0, 1.0, 0.2)
Y = np.arange(-1.0, 1.0, 0.2)
```

```python
Z = np.zeros((10, 10))
```

```python
#重み
w_x = 2.5
w_y = 3.0

bias = 0.1
```

```python
for i in range(10):
    for j in range(10):
        
        u = X[i]*w_x + Y[j]*w_y + bias
        
        y = 1/(1+np.exp(-u))
        Z[j][i] = y
```

```python
Z
```

```python
plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

<h3>ニューラルネットワークの実装</h3>

```python
def middle_layer(x, w, b):
    u = np.dot(x, w) + b
    return 1/(1+np.exp(-u))
```

```python
def output_layer(x, w, b):
    u = np.dot(x, w) + b
    return u
```

```python
w_im = np.array([[4.0, 4.0],
                 [4.0, 4.0]])
w_mo = np.array([[1.0], [-1.0]])
```

```python
b_im = np.array([3.0, -3.0])
b_mo = np.array([0.1])
```

```python
X = np.arange(-1.0, 1.0, 0.2)
Y = np.arange(-1.0, 1.0, 0.2)

Z = np.zeros((10,10))
```

```python
for i in range(10):
    for j in range(10):
        
        #順伝播
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)
        
        Z[j][i] = out[0]
```

```python
plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

<h2>ニューラルネットワーク（分類）</h2>

```python
X = np.arange(-1.0, 1.0, 0.1)
Y = np.arange(-1.0, 1.0, 0.1)

w_im = np.array([[1.0, 2.0],
                 [2.0, 3.0]])
w_mo = np.array([[-1.0, 1.0],
                 [ 1.0,-1.0]])

b_im = np.array([0.3, -0.3])
b_mo = np.array([0.4, 0.1])
```

```python
def middle_layer(x, w, b):
    u = np.dot(x, w) + b
    return 1/(1+np.exp(-u))
```

```python
def output_layer(x, w, b):
    u = np.dot(x, w) + b
    return np.exp(u)/np.sum(np.exp(u)) #softmax
```

```python
x_1 = []
y_1 = []
x_2 = []
y_2 = []
```

```python
for i in range(20):
    for j in range(20):
        #順伝播
        inp = np.array([X[i], Y[j]])
        mid = middle_layer(inp, w_im, b_im)
        out = output_layer(mid, w_mo, b_mo)
        
        if out[0] > out[1]:
            x_1.append(X[i])
            y_1.append(Y[j])
        else:
            x_2.append(X[i])
            y_2.append(Y[j])
```

```python
plt.scatter(x_1, y_1, marker='+')
plt.scatter(x_2, y_2, marker='o')
plt.show()
```
